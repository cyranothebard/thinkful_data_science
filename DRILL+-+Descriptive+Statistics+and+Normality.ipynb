```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create CHI SQUARED distribution
chisq = np.random.chisquare(3, 100)

#Plot a histogram for rand1.
plt.hist(chisq, bins=20, color='c')

# Add a vertical line at the mean.
plt.axvline(chisq.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(chisq.mean() + chisq.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(chisq.mean() - chisq.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
"""Observation: The close a distribution is to Normal the more useful are the mean and stdev to give useful information
about the data set."""
```

```python
norm1 = np.random.normal(5, 0.5, 100)
norm2 = np.random.normal(10, 1, 100)
norm3 = norm1 + norm2

plt.hist(norm3, bins=10, color = 'c')
plt.axvline(norm3.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(norm3.mean() + norm3.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(norm3.mean() - norm3.std(), color = 'b', linestyle = 'dashed', linewidth = 2)

plt.show()
```

```python
norm1.sort()
norm2.sort()
add_rand = np.random.normal(15, 1.5, 100)

plt.plot(norm1, norm3, "o")
plt.title('norm1 vs norm3, qq plot')
plt.show()

plt.plot(norm1, norm2, "o")
plt.title('norm1 vs norm2, qq plot')
plt.show()

plt.plot(norm3, add_rand, "o")
plt.title('norm3 vs random normal set')
plt.show()
```

```python
"""Observation: This additive property doesn't apply to a normal distribution. The histogram of norm3 does not have 
the requisite shape of a normal distribution. This is reinforced by the QQ plots showing that, while norm1 vs norm2 
is roughly linear, both norm1 vs norm3 and norm3 vs random normal set are random distributions."""
```
