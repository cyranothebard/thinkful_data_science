```python
I decided to work with the Kaggle Election Finanace data set available at the following link: https://www.kaggle.com/danerbland/electionfinance
In this data set I wanted to explore three questions.
(1) Was there a positive correlation between money spent and election outcome?
(2) How did the strength of the effect differ for encumbents vs challengers?
(3) Was this effect stronger for Republicans or Democrats?
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/brandonlewis/Desktop/Data Science with Thinkful/CandidateSummaryAction1.csv')
Candidate_Data = pd.DataFrame(df)
print(Candidate_Data)
```

```python
In the first figure I am trying to see if there is a substantive difference in net contributions between winning 
and loosing candidates. I have plotted two lines, one for net contributions of winning candidates and one for 
net contributions of loosing candidates.
```

```python
plt.plot(lambda Candidate_Data: Candidate_Data['net_con'] if Candidate_Data['winner'] == 'Y', color = 'purple')
plt.plot(Candidate_Data.ix[lambda: Candidate_Data['net_con'] if Candidate_Data['winner'] == 'N'], color = 'orange')
plt.ylabel('Net Contributions')
plt.title('Net Contribution for Winning Candidates')

plt.show()
```

```python
Next, I wanted to explore other possible dividing lines in this data. So I decideed to plot net contributions for winning 
encumbents versus winning challengers. 
```

```python
plt.plot(lambda Candidate_Data: Candidate_Data['net_con'] if Candidate_Data['winner'] == 'Y' 
         and Candidate_Data['can_inc_cha_ope_sea'] == 'INCUMBENT', color = 'purple')
plt.plot(lambda Candidate_Data: Candidate_Data['net_con'] if Candidate_Data['winner'] == 'Y' 
        and Candidate_Data['can_inc_cha_ope_sea'] == 'CHALLENGER', color = 'orange')
plt.ylabel('Net Contributions')
plt.title('Contribution Difference for Challenger vs Incumbent')
plt.show()
```

```python
Finally, to compare the effect on Republicans versus Democrats, I plotted net contribution for winning candidates
as two separate lines for Democrats and Republicans.
```

```python
plt.plot(lambda Candidate_Data: Candidate_Data['net_con'] if Candidate_Data['winner'] == 'Y' 
         and Candidate_Data['can_par_aff'] == 'REP', color = 'purple')
plt.plot(lambda Candidate_Data: Candidate_Data['net_con'] if Candidate_Data['winner'] == 'Y' 
         and Candidate_Data['can_par_aff'] == 'DEM', color = 'orange')
plt.ylabel('Net Contributions')
plt.title('Net Contribution for Winning Candidates by Party')
plt.show()
```
